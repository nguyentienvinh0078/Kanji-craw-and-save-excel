```python
from Driver import Driver
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from time import sleep
from util import json_write, json_read
```

```python
the_driver = Driver()
driver = the_driver.init_driver()
driver.get('https://www.tieng-nhat.com/2015/08/2000-chu-kanji-thong-dung-nhat.html')
```

```python
row_elements = driver.find_elements(By.XPATH, '/html/body/div[6]/div[2]/div/div[2]/div[1]/div/article/div[3]/table/tbody/tr')
kanji_2000_data = {}
title = ['index', 'kanji_214', 'vn_sound', 'mean']
for i in range(2, len(row_elements) + 1):
    row_data = {}
    title = ['index', 'word', 'vn_sound', 'mean', 'jp_sound_char']
    col_elements = driver.find_elements(By.XPATH, f'/html/body/div[6]/div[2]/div/div[2]/div[1]/div/article/div[3]/table/tbody/tr[{i}]/th')
    for j in range(1, len(col_elements) + 1):
        col_element = driver.find_element(By.XPATH, f'/html/body/div[6]/div[2]/div/div[2]/div[1]/div/article/div[3]/table/tbody/tr[{i}]/th[{j}]')
        
        row_data.update({
            f"{title[j-1]}": col_element.text
        })
    
    word_data = {
        f"{row_data['word']}": {
            "index": row_data['index'],
            "vn_sound": row_data['vn_sound'],
            "mean": row_data['mean'],
            "jp_sound_char": row_data['jp_sound_char']
        }
    }

    print(row_data)

    kanji_2000_data.update(word_data)
    json_write('Kanji Database\\kanji_2000.json', {
        'kanji_2000': kanji_2000_data
    })
```

```python
driver.quit()
```
